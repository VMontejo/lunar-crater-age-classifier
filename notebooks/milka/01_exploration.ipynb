```python
from pathlib import Path

data_dir = Path("/home/milka/code/VMontejo/lunar-crater-age-classifier/raw_data/7041842")
data_dir, data_dir.exists()
```

```python
list(data_dir.iterdir())
```

```python
import zipfile

for zip_path in data_dir.glob("*.zip"):
    print("Extracting:", zip_path.name)
    extract_dir = data_dir / zip_path.stem   # e.g. ejecta_train/
    extract_dir.mkdir(exist_ok=True)
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall(extract_dir)

print("Done!")
```

```python
for p in data_dir.iterdir():
    if p.is_dir():
        print(p.name, "->", len(list(p.iterdir())), "items")
```
